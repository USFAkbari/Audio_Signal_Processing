<h1>Zero Crossing Rate

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import pandas as pd

from glob import glob
import librosa
import librosa.display
```

<h2>Reading audio files

```python
audio_files = glob('D:\GitHub_Repository\Audio_Signal_Processing\Sound\*')
audio_files
```

```python
#play audio file 
ipd.Audio(audio_files[1])
```

```python
y, sr = librosa.load(audio_files[1])
print(f'Sample Rate :{sr}')
```

```python
pd.Series(y).plot(figsize=(14,5),
                                    lw=1 ,
                                    title ='Raw Audio Example')
plt.show()
```

```python
y_trimmed , _ =librosa.effects.trim(y, top_db=20)

pd.Series(y_trimmed).plot(figsize=(14,5),
                                    lw=1 ,
                                    title ='Raw Audio Trimmed Example')
plt.show()
```

```python
pd.Series(y[0:500]).plot(figsize=(14,5),
                                    lw=1 ,
                                    title ='Raw Audio Zoomed Example')
plt.show()
```

```python
zero_crossings = librosa.zero_crossings(y[0:500], pad=False)
```

```python
zero_crossings.shape
```

```python
print(sum(zero_crossings))
```
